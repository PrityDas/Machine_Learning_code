```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
titanic=sns.load_dataset('titanic')
```

```python
tips.head()
```

```python
sns.set(style="darkgrid")
sns.relplot(x='total_bill',y='tip', hue='time', data=tips)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.relplot(x='total_bill',y='tip', kind='line',  data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker', data=tips)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title("Rehreesion Line")
```

```python
tips.head()
```

```python
corr=tips.corr()# pearson coeffiecnt correlation #chi square correlation
```

```python
sns.heatmap(corr, annot=True, cmap=plt.cm.CMRmap_r)
```

```python
sns.kdeplot(x='total_bill', data=tips, shade=True, color='green') #KDE-- Kernel Density Estimation
```

```python
sns.kdeplot(x='size', data=tips, shade=True, color='green')
```

```python
sns.kdeplot(x='tip', data=tips, shade=True, color='green')
```
